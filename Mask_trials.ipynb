Load annotations


```python
import json

def load_annotations(annotation_file):
    with open(annotation_file, 'r') as f:
        data = json.load(f)

    # Create a mapping from image_id to filename
    image_id_to_filename = {img['id']: img['file_name'] for img in data['images']}

    annotations = {}
    for ann in data['annotations']:
        image_id = ann['image_id']
        filename = image_id_to_filename.get(image_id)  # Lookup the filename from image_id

        if image_id not in annotations:
            annotations[image_id] = {'filename': filename, 'annotations': []}

        annotations[image_id]['annotations'].append({
            'segmentation': ann['segmentation'],
            'orientation': ann['attributes'].get('Orientation', None),
            'category': ann['category_id']
        })

    return annotations
# Load the JSON file

annotations = load_annotations(annotations_path)

# Check loaded annotations
for image_id, data in annotations.items():
    print(f"Image ID: {image_id}, Filename: {data['filename']}, Annotations: {len(data['annotations'])}")
```

Ensure all validations are present and in the correct format

```python
def validate_annotations(annotations):
    errors = []
    for image_id, data in annotations.items():
        filename = data['filename']  # Access the filename for each image_id
        annotation_list = data['annotations']  # Access the list of annotations for the image_id

        for idx, ann in enumerate(annotation_list):
            error_msg = f"Error in Image ID {image_id} ({filename}), Annotation {idx + 1}:"
            has_error = False

            # Check if segmentation is present and valid
            if 'segmentation' not in ann or not ann['segmentation'] or len(ann['segmentation'][0]) < 6:
                has_error = True
                error_msg += " Invalid or missing segmentation."

            # Check if orientation is present and within the correct range
            if 'orientation' not in ann or not (0 <= ann['orientation'] < 360):
                has_error = True
                error_msg += " Invalid or missing orientation."

            # Check if category ID is present and valid
            if 'category' not in ann or not isinstance(ann['category'], int):
                has_error = True
                error_msg += " Invalid or missing category ID."

            # Log the error if any validation fails
            if has_error:
                errors.append(error_msg)

    if errors:
        print("\n".join(errors))
    else:
        print("All annotations are valid!")

# Run the validation on the updated annotations
validate_annotations(annotations)
```

```python

import cv2
import os

def load_images(image_folder, annotations):
    images = {}
    # Iterate over the annotations
    for image_id, data in annotations.items():
        filename = data['filename']  # Get the filename associated with the image_id

        # Construct the full path to the image using the filename (no need for image_id anymore)
        image_file = os.path.join(image_folder, filename)

        if os.path.exists(image_file):
            image = cv2.imread(image_file)
            images[image_id] = image
            print(f"Image file found: {image_file}")
        else:
            print(f"Image file not found: {image_file}")

    return images

# Load the images using the updated function
images = load_images(image_folder_path, annotations)

# Check the loaded images
print("Loaded Images:", list(images.keys()))
```

```python
import numpy as np
import cv2

def create_masks(image_shape, annotation_list):
    masks = []
    for ann in annotation_list:
        mask = np.zeros(image_shape[:2], dtype=np.uint8)
        segmentation = ann['segmentation']

        # Flatten segmentation if it's nested
        if len(segmentation) == 1 and isinstance(segmentation[0], list):
            segmentation = segmentation[0]

        # Ensure segmentation is a flat list
        if isinstance(segmentation, list):
            polygon = np.array(segmentation, dtype=np.int32).reshape(-1, 2)

            if polygon.shape[0] >= 3:  # Valid polygon
                cv2.fillPoly(mask, [polygon], 255)
                masks.append((mask, ann['orientation'], ann['category']))
            else:
                print(f"Invalid polygon with less than 3 points: {segmentation}")
        else:
            print(f"Invalid segmentation format for image: {segmentation}")
    return masks

# Test the updated function
test_image_name = list(images.keys())[0]  # Pick one image ID (filename in this case)
test_image = images[test_image_name]
test_annotations = annotations[test_image_name]  # Fetch annotations by the image name

masks = create_masks(test_image.shape, test_annotations)

# Check masks
print(f"Created {len(masks)} masks for image {test_image_name}")
```
